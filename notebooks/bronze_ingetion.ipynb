Loading 7 csv files as Delta tables without applying any transformations.

```python
from pyspark.sql.functions import current_timestamp, lit
from pyspark.sql import SparkSession
```

```python
#initializing spark
spark = SparkSession.builder.getOrCreate()
```

```python
#setting the location for raw data 
BASE_PATH = 'Files/raw/adventureworks'

#mapping tables names to respectice csv files 
tables = {
    'bronze_sales_order_header': 'sale_order_header.csv',
    'bronze_sales_order_detail': 'sales_order_detail.csv',
    'bronze_product': 'product.csv',
    'bronze_product_category': 'product_category.csv',
    'bronze_product_subcategory': 'product_subcategory.csv',
    'bronze_customer': 'customer.csv',
    'bronze_sales_territory': 'sales_territory.csv'
}

#saving csv files as delta tables
for table_name, filename in tables.items():
    df = spark.read.option('header', True).option('inferSchema', True).csv(f'{BASE_PATH}/{filename}')
    df = df.withColumn('_ingested_at', current_timestamp()).withColumn('_source_file', lit(filename))
    df.write.format('delta').mode('overwrite').option('overwriteSchema', 'true').saveAsTable(table_name)
    print(f'{table_name}: {df.count()} rows')

print('Bronze ingestion complete')
```
